```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.font_manager as fm
plt.rc('font', family = 'Malgun Gothic')
```

```python
df = pd.read_csv("06. 데이터 시각화 기초_penguins_lter_result.csv")
```

```python
# 1.Bar Graph
```

```python
species_counts = df['Species'].value_counts()
```

```python
# Basic drawing of bar graph
plt.figure(figsize = (7, 5))

sns.barplot(x = species_counts.index, y = species_counts.values)

plt.title('Penguins specie counts')
plt.xlabel('Species')
plt.ylabel('Count')

plt.show()
```

```python
# 2.Heatmap Graph
```

```python
num_data = df.select_dtypes(include = ['number'])
corr_matrix = num_data.corr()
print(corr_matrix)
```

```python
# Basic drawing of heatmap graph
plt.figure(figsize = (7, 5))

sns.heatmap(corr_matrix, cmap = 'Blues', linewidths = 0.5, annot = True)

plt.title('Heatmap of Penguin Characteristics')
```

```python
# 3.Scatter Graph
```

```python
# Basic drawing of scatter graph
plt.figure(figsize = (7, 5))

sns.scatterplot(data = df, x = 'Flipper Length (mm)', y = 'Body Mass (g)')

plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.title('Penguin Flipper Length vs Body Mass')

plt.show()
```

```python
# Decorating a scatter graph
plt.figure(figsize = (7, 5))

sns.scatterplot(data = df, x = 'Flipper Length (mm)', y = 'Body Mass (g)', hue = 'Species', palette = 'muted')

plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.title('Penguin Flipper Length vs Body Mass')

plt.show()
```

```python
# 4.Line Graph
```

```python
df['Date Egg'] = pd.to_datetime(df['Date Egg'], errors='coerce')
df['Date Egg'].head()
```

```python
date_counts = df.groupby(['Date Egg']).size().reset_index(name='Count')
```

```python
# Basic drawing of line graph
plt.figure(figsize = (5, 5))

sns.lineplot(x = 'Date Egg', y = 'Count', data = date_counts)

plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.title('Penguin Flipper Length vs Body Mass')

plt.show()
```
